```python
#automodelforcausalLM to download the model
#autoTokenizer to get the tokenizer
#set_seed to regenerate something generated before
from transformers import AutoModelForCausalLM, AutoTokenizer, set_seed
```

```python
#convert tensor to cuda  because it's hard for to download a huge model(allocate in gpu memory)
import torch
torch.set_default_tensor_type(torch.cuda.FloatTensor)
```

```python
model = AutoModelForCausalLM.from_pretrained("bigscience/bloom-1b3", use_cache=True)
tokenizer = AutoTokenizer.from_pretrained("bigscience/bloom-1b3")
```

```python
#reproduce the test
set_seed(424242)
```

```python
model.__class__.__name__
```

```python
blogpost_title = 'Financial freedom'
```

```python

prompt = f'This is a blog post on {blogpost_title} \n {blogpost_title} will change your life. '
```

```python
prompt
```

```python
input_ids = tokenizer(prompt, return_tensors="pt").to(0)
```

```python
#top_k and temperature helps to generate diverse content and repetition_penalty reduce repetition
sample = model.generate(**input_ids, max_length=300,  top_k=1, temperature=0.9, repetition_penalty = 2.0)
```

```python
print(tokenizer.decode(sample[0], truncate_before_pattern=[r"\n\n^#", "^'''", "\n\n\n"]))
```

```python
#num_beam and num_beam_groups search many differnt angles to create content

sample = model.generate(**input_ids, max_length=300, num_beams = 2, num_beam_groups = 2, top_k=1, temperature=0.9, repetition_penalty = 2.0, diversity_penlty=0.9)
```

```python
print(tokenizer.decode(sample[0], truncate_before_pattern=[r"\n\n^#", "^'''", "\n\n\n"]))
```
